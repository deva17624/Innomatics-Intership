## Maximum Product of Two Elements in an Array

```python
class Solution(object):
    def maxProduct(self, nums):
        nums.sort()
        return (nums[-1]-1)*(nums[-2]-1)
```

```python
sol=Solution()
sol.maxProduct([3,4,5,2])
```

## Count Number of Teams

```python
class Solution(object):
    def numTeams(self, rating):
        rat=rating
        total = 0
        n = len(rat)

        for mid in range(n):
            left_small = 0
            left_large = 0
            right_small = 0
            right_large = 0

            for i in range(mid):
                if rat[i] < rat[mid]:
                    left_small += 1
                else:
                    left_large += 1

            for k in range(mid + 1, n):
                if rat[k] > rat[mid]:
                    right_large += 1
                else:
                    right_small += 1

            total += left_small * right_large
            total += left_large * right_small

        return total
```

```python
sol=Solution()
sol.numTeams([2,5,3,4,1])
```

## Number of Students Doing Homework at a Given Time

```python
class Solution(object):
    def busyStudent(self, startTime, endTime, queryTime):
        s=0
        for i in range(len(startTime)):
            if startTime[i] <= queryTime <= endTime[i]:
                s=s+1
        return s
```

```python
sol=Solution()
sol.busyStudent([1,2,3], [3,2,7],  4)
```

## Number of Steps to Reduce a Number to Zero

```python
class Solution(object):
    def numberOfSteps(self, num):
        n=0
        while(num!=0):
            if num%2==0:
                num=num/2
            else:
                num=num-1
            n=n+1
        return n
        
```

```python
sol=Solution()
sol.numberOfSteps(14)
```

##  Counting Bits

```python
class Solution(object):
    def countBits(self, n):
        ans = [0] * (n + 1)
        for i in range(1, n + 1):
            ans[i] = ans[i >> 1] + (i & 1)
        return ans
```

```python
sol=Solution()
sol.countBits(2)
```
