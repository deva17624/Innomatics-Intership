### 1 Running Sum of 1d Array

```python
class Solution(object):
    def runningSum(self, nums):
        l=[]
        s=0
        for i in nums:
            s=s+i
            l.append(s)
        return l
```

```python
sol = Solution()
print(sol.runningSum([1,2,3,4]))
```

### 2.Shuffle the Array

```python
class Solution(object):
    def shuffle(self, nums, n):
        l1=nums[:n]
        l2=nums[n:]
        l=[]
        for i in range(n):
            l.append(l1[i])
            l.append(l2[i])
        return l
```

```python
sol = Solution()
print(sol.shuffle([2,5,1,3,4,7],3))
```

### 3 Kids With the Greatest Number of Candies

```python
class Solution(object):
    def kidsWithCandies(self, candies, extraCandies):
        m=max(candies)
        l=[]
        for i in candies:
            s=i+extraCandies
            if m<=s:
                l.append(True)
            else:
                l.append(False)

        return l
```

```python
sol = Solution()
print(sol.kidsWithCandies([2,3,5,1,3],3))
```
