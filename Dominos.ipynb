```python
df["order_placed_at"] = pd.to_datetime(df["order_placed_at"])
df["order_delivered_at"] = pd.to_datetime(df["order_delivered_at"])
```

```python
df.dtypes
```

```python
df["delivery_min"] = (
    df["order_delivered_at"] - df["order_placed_at"]
).dt.total_seconds() / 60
```

```python
df[["order_id", "delivery_min"]].head()
```

```python
df["delivery_min"].describe()
```

```python
val= df["delivery_min"].quantile(0.95)
val
```

```python
if val <= 31:
    print("Store meets Diminos 95th percentile requirement")
else:
    print(" Store does NOT meet Diminos requirement")
```

```python
late_orders = df[df["delivery_min"] > 31]
late_orders.shape
```
